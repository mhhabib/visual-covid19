```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
cov=pd.read_csv("covid.csv")
cov
```

```python
aa=cov.drop(columns=['Province/State'])
bb=aa.drop(columns=['Lat'])
cc=bb.drop(columns=['Long'])
cc.to_csv('covi19.csv',index=False)
```

```python
covv=cov.sort_values('3/30/20',ascending=False)
print(covv[['Country/Region','3/30/20']][0:30])
```

```python
abc=covv[['Country/Region','3/30/20']][0:15]
final=abc.sort_values('3/30/20',ascending=True)
plt.barh(final['Country/Region'],final['3/30/20']/10**3)
plt.xlabel("Infected ->",color='red',)
plt.ylabel("Country ->",color='red')
plt.title("Covid19 Epedemic",color='red')
plt.show()
```

```python
took=pd.read_csv("covi19.csv")
took
```

```python
took.groupby(['Country/Region']).sum()
```

```python
tkk=took.groupby(['Country/Region']).sum()
tkk.sort_values('3/30/20',ascending=False)
tkk.to_csv('covifinal.csv')
#print(tk[['Country/Region','3/30/20']][0:30])
```

```python
df=pd.read_csv("covifinal.csv")
dff=df.sort_values('3/30/20',ascending=False)
print(dff[['Country/Region','3/30/20']][0:30])
```

```python
dfff=dff[['Country/Region','3/30/20']][0:15]
df_final=dfff.sort_values('3/30/20',ascending=True)
plt.barh(df_final['Country/Region'],df_final['3/30/20']/10**3)
plt.xlabel("Coronavirus Cases in Thousand",fontsize="16")
plt.title("Covid19 Epedemic",color='blue',fontsize="20")
plt.show()
```

```python
df=pd.read_csv("death.csv")
aa=df.drop(columns=['Province/State'])
bb=aa.drop(columns=['Lat'])
cc=bb.drop(columns=['Long'])
cc.to_csv('modi_death.csv',index=False)
```

```python
df=pd.read_csv('modi_death.csv')
tkk=df.groupby(['Country/Region']).sum()
tkk.sort_values('3/30/20',ascending=False)
tkk.to_csv('final_death.csv')
```

```python
df=pd.read_csv('final_death.csv')
dff=df.sort_values('3/30/20',ascending=False)
print(dff[['Country/Region','3/30/20']][0:30])
```

```python
dfff=dff[['Country/Region','3/30/20']][0:15]
df_final=dfff.sort_values('3/30/20',ascending=True)
plt.barh(df_final['Country/Region'],df_final['3/30/20'],color="red")
plt.xlabel("Passed away",fontsize="16")
plt.title("Covid19 Epedemic",color='red',fontsize="20")
plt.show()
```
